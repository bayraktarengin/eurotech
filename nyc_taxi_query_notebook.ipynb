```python
  # 2. Most expensive trip per mile (Total_amount/Trip_distance).
q = """
        SELECT *, (Total_amount/Trip_distance) as cost_per_mile
        FROM nyc_data_sep
        ORDER BY cost_per_mile DESC
        LIMIT 1;
    """
sqldf(q)
```

```python
 # 4. Longest trip duration.
 # strftime



q = """
        SELECT (CAST(strftime('%s', Lpep_dropoff_datetime) as integer)-CAST(strftime('%s', Lpep_pickup_datetime) as integer)) as trip_duration
        FROM nyc_data_sep
        ORDER BY trip_duration DESC
        LIMIT 1;
    """
sqldf(q)
```

   7. Average total trip by day of week (Fortunately, we have day of week information. Otherwise, we need to create a new date column without hours from date column. Then, we need to create "day of week" column, i.e Monday, Tuesday .. or 1, 2 ..,  from that new date column. Total trip count should be found for each day, lastly average total trip should be calculated for each day).

```python
nyc_data_sep.head()
```

```python
q = """
        SELECT strftime('%Y%m%d', Lpep_pickup_datetime) as date,
               lpep_pickup_day_of_week,
               count(*)
        FROM nyc_data_sep
        group by date
  
        
        
    """
sqldf(q)

```

```python
q = """
      SELECT * 
      FROM
           ( SELECT strftime('%Y%m%d', Lpep_pickup_datetime) as date,
                    lpep_pickup_day_of_week,
                    count(*) as day_trip_count
           FROM nyc_data_sep
           group by date)
  
        
        
    """
sqldf(q)
```

```python
q = """
      SELECT lpep_pickup_day_of_week,avg(day_trip_count)
      FROM
           ( SELECT strftime('%Y%m%d', Lpep_pickup_datetime) as date,
                    lpep_pickup_day_of_week,
                    count(*) as day_trip_count
           FROM nyc_data_sep
           group by date)
      group by lpep_pickup_day_of_week
        
        
    """
sqldf(q)
```

```python
q = """
      SELECT lpep_pickup_day_of_week,count(*)
      FROM
           ( SELECT strftime('%Y%m%d', Lpep_pickup_datetime) as date,
                    lpep_pickup_day_of_week,
                    count(*) as day_trip_count
           FROM nyc_data_sep
           group by date)
        group by lpep_pickup_day_of_week
  
        
        
    """
sqldf(q)
```

```python
#case  when ile day_of_week i dÃ¼zelt
q = """

      SELECT day_of_week,avg(day_trip_count)
                                        
      FROM
           ( SELECT strftime('%Y%m%d', Lpep_pickup_datetime) as date,
                    strftime('%w', Lpep_pickup_datetime) as day_of_week,
                    count(*) as day_trip_count
           FROM nyc_data_sep
           group by date)
      group by day_of_week
        

        
    """
sqldf(q)
```
