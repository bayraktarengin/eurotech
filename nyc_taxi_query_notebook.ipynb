```python
 #  1. Most expensive trip (total amount).
q = """
        SELECT total_amount 
        FROM nyc_data_sep
        ORDER BY total_amount DESC
        LIMIT 1;
    """

sqldf(q)
```

```python
q = """
        SELECT * 
        FROM nyc_data_sep
        ORDER BY total_amount DESC
        LIMIT 1;
    """

sqldf(q)
```

```python
  # 2. Most expensive trip per mile (Total_amount/Trip_distance).
q = """
        SELECT *, (Total_amount/Trip_distance) as cost_per_mile
        FROM nyc_data_sep
        ORDER BY cost_per_mile DESC
        LIMIT 1;
    """
sqldf(q)
```

```python
q = """
        SELECT MAX (Total_amount/Trip_distance) as cost_per_mile
        FROM nyc_data_sep
        
    """
sqldf(q)
```

```python
 #  3. Most generous trip (highest tip).

q = """
        SELECT *
        FROM nyc_data_sep
        ORDER BY Tip_amount DESC
        LIMIT 1;
    """
sqldf(q)
```

```python
q = """
        SELECT MAX (Tip_amount) 
        FROM nyc_data_sep
        
    """
sqldf(q)
```

```python
 # 4. Longest trip duration.
 # strftime



q = """
        SELECT (CAST(strftime('%s', Lpep_dropoff_datetime) as integer)-CAST(strftime('%s', Lpep_pickup_datetime) as integer)) as trip_duration
        FROM nyc_data_sep
        ORDER BY trip_duration DESC
        LIMIT 1;
    """
sqldf(q)
```

```python
q = """
        SELECT DATEDIFF ('Second', 'Lpep_dropoff_datetime', 'Lpep_pickup_datetime')
        FROM nyc_data_sep
        LIMIT 1;
    """
sqldf(q)
```
