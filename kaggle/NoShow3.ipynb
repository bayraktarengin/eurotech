```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("no_show.csv")
df.head()
```

```python
df["ScheduledDaydt"]=pd.to_datetime(df["ScheduledDay"])
df["AppointmentDaydt"]=pd.to_datetime(df["AppointmentDay"])
```

```python

df["DayDifference"] = (df["AppointmentDaydt"] - df["ScheduledDaydt"]).dt.days
```

```python
df.head()
```

```python
df["DayDifference"].value_counts()
```

```python
df[df["DayDifference"]==1]
```

```python
df["DayDifference"]=df["DayDifference"]+1
```

```python
df.head()
```

```python
yaslar=[-1,2,12,18,30,45,60,120]
label=["bebek","cocuk","ergen","genc","olgun","ortayas","yasli"]
df["AgeCat"]=pd.cut(df["Age"],bins=yaslar,labels=label)
```

```python
bins_day=[0,10,30,60,120,180]
labels_day=["0-10","10-30","30-60","60-120","120-180"]

df["DayDifference"]=pd.cut(df["DayDifference"],bins=bins_day, labels=labels_day)
```

```python
catcols=['Gender', 'Neighbourhood', 'Scholarship', 'Hipertension', 'Diabetes', 'Alcoholism', 'Handcap', 'SMS_received','DayDifference', 'AgeCat']
```

```python
df.columns
```

```python
plt.figure(figsize=(12, 20))
for i, col in enumerate (catcols,1):
    plt.subplot(int(len(catcols)/2),2,i)
    sns.countplot(x=col,data=df,hue="No-show")
    plt.xticks(rotation=90)
plt.show()
```
