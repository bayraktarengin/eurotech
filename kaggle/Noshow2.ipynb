```python
import pandas as pd
```

```python
df=pd.read_csv("no_show.csv")
df.head()
```

```python
df.duplicated().sum()
```

```python
df['PatientId'].nunique()
```

```python
df.loc[df['PatientId'].duplicated(keep=False)].sort_values('PatientId')
```

```python
df.duplicated(['PatientId','No-show']).sum()
```

```python
df.drop_duplicates(subset=['PatientId','No_Show'],inplace=True)
```

```python
df.groupby('PatientId')['No_Show'].sum()
```

```python
df_path_duplicated=df.loc[df['PatientId'].duplicated(keep=False)]
```

```python
df_pat_total_count=df_path_duplicated.groupby(['PatientId'], as_index=False).size()
df_pat_total_count.head()
```

```python
df_pat_total_count.rename(columns={'size':'total'},inplace=True)
```

```python
df_pat_NoShowCount=df_path_duplicated.groupby(['PatientId','No-show'], as_index=False).size()
```

```python
df_pat_NoShowCount.head()
```

```python
df_pat_total_final=pd.merge(df_pat_total_count,df_pat_NoShowCount,on='PatientId')
```

```python
df_pat_total_final.head()
```

```python
df_pat_total_final['count_pst']=df_pat_total_final['size']/df_pat_total_final['total']
```
